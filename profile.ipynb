```python
from os import system, environ
from pathlib import Path
environ["PROFILING"] = "1"
models = [x.stem for x in list(Path('.').glob('./examples/*.py'))]
for model in models:
    system(f"python -m cProfile -o ./examples/profiles/{model}.prof ./examples/{model}.py")
```

```python
import pstats


def profile(model: str) -> None:
    p = pstats.Stats(f"./examples/profiles/{model}.prof")
    p.sort_stats("cumulative").print_stats(50)


for model in models:
    profile(model)
```
