```python
from os import system


for model in ["linear", "mlp"]:
    system(f"python -m cProfile -o ./profiles/{model}.prof -m cudagrad.{model}")
system("git restore cudagrad/plots/*")
```

```python
import pstats


def profile(model: str) -> None:
    p = pstats.Stats(f"./profiles/{model}.prof")
    p.sort_stats("cumulative").print_stats(100)
```

```python
profile("linear")
```

```python
profile("mlp")
```
