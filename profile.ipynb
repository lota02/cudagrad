```python
from os import system, environ

environ["PROFILING"] = "1"
for model in ["linear", "mlp", "moons"]:
    system(f"python -m cProfile -o ./profiles/{model}.prof -m cudagrad.{model}")
```

```python
import pstats


def profile(model: str) -> None:
    p = pstats.Stats(f"./profiles/{model}.prof")
    p.sort_stats("cumulative").print_stats(50)
```

```python
profile("linear")
```

```python
profile("mlp")
```

```python
profile("moons")
```
